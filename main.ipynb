```python
import pandas as pd
df1 = pd.read_csv('/content/extracted_data.csv')
df1.head()
```

```python
df_cleaned = text_cleaner.clean_dataframe(df1, text_column='Data')
df_cleaned = text_cleaner.clean_dataframe(df_cleaned, text_column='Content')
df_cleaned.head()
```

```python
from summarising_news import NewsSummarizer
model_checkpoint_path = '/content/drive/MyDrive/text_summarization_model_checkpoint.ckpt'
news_summarizer = NewsSummarizer(model_checkpoint_path)
```

```python
df_with_summaries = news_summarizer.summarize_dataframe(df_cleaned, text_column='Data', summary_column='summary')
```

```python
df_with_summaries.head()
```

```python
from classifying_news import ModelPredictor
```

```python
classifying = ModelPredictor(model_path = '/content/drive/MyDrive/Text_Classification_model.joblib')
df3 = classifying.predict_and_add_column(df_with_summaries, input_column = 'summary', output_column = 'Label')
```

```python
df3.head()
```

```python
df3.to_csv('final_data.csv')
```

```python
# Group by classification
grouped = df3.groupby('Label')

# Create and write to a text file
with open('news_summary.txt', 'w') as file:
    for category, group in grouped:
        file.write(f"Category: {category.capitalize()}\n")
        for index, row in group.iterrows():
            file.write(f"Headline: {row['Headline']}\n")
            file.write(f"Summary: {row['summary']}\n")
            file.write("\n")
        file.write("="*50 + "\n")
        file.write("\n")

print("Text file 'news_summary.txt' has been generated.")
```
